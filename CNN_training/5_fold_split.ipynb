```python
# Here we take adj data and create a 5 fold cross vlaidaiton split. This means the output is 5 pairs of test and train files.
# This will beused to get an error on evaultion. Evaluation of the network will take place on the entire testing set. 
from sklearn.model_selection import KFold
from prody import *
from biopandas.pdb import PandasPdb
import numpy as np
from scipy.spatial import distance_matrix
import matplotlib.pyplot as plt
import os
import pandas as pd
import csv 
from sklearn.preprocessing import KBinsDiscretizer
from sklearn.model_selection import train_test_split
```

```python

def five_fold_splits(full_set,var_name):
    '''This function saves csv 5 fold corss validation train and test sets for CNN binary classification'''
    full_set = full_set.dropna() # drop nans..
    # Even out class names so over all df the classes are equaly split.
    binding = full_set[[1]]
    expr = full_set[[2]]
    est = KBinsDiscretizer(n_bins=2, encode='ordinal', strategy='quantile')
    est.fit(binding)
    binding = est.transform(binding)
    est = KBinsDiscretizer(n_bins=2, encode='ordinal', strategy='quantile')
    est.fit(expr)
    expr = est.transform(expr)
    full_set[[1]] = binding
    full_set[[2]] = expr

    kf = KFold(n_splits=5) 
    for id, (train_index, test_index) in enumerate(kf.split(full_set)):
        train, test = full_set.iloc[train_index], full_set.iloc[test_index]
        # assert each train and test are different..
        train.to_csv("/mnt/ncshare/ozkilim/charge_pca_deepmut/CNN_training/cross_val_splits/" + var_name + "_train_" + str(id) + '.csv')
        test.to_csv("/mnt/ncshare/ozkilim/charge_pca_deepmut/CNN_training/cross_val_splits/" + var_name + "_test_" + str(id) + '.csv')


# For each dataset, create 5 fold splits for CNN training.

full_set = pd.read_csv("/mnt/ncshare/ozkilim/charge_pca_deepmut/CNN_training/wuhan_adj_data.csv",header=None)
var_name = 'wuhan'
five_fold_splits(full_set,var_name)

full_set = pd.read_csv("/mnt/ncshare/ozkilim/charge_pca_deepmut/CNN_training/alpha_adj_data.csv",header=None)
var_name = 'alpha'
five_fold_splits(full_set,var_name)

full_set = pd.read_csv("/mnt/ncshare/ozkilim/charge_pca_deepmut/CNN_training/beta_adj_data.csv",header=None)
var_name = 'beta'
five_fold_splits(full_set,var_name)

full_set = pd.read_csv("/mnt/ncshare/ozkilim/charge_pca_deepmut/CNN_training/delta_adj_data.csv",header=None)
var_name = 'delta'
five_fold_splits(full_set,var_name)

full_set = pd.read_csv("/mnt/ncshare/ozkilim/charge_pca_deepmut/CNN_training/eta_adj_data.csv",header=None)
var_name = 'eta'
five_fold_splits(full_set,var_name)

full_set = pd.read_csv("/mnt/ncshare/ozkilim/charge_pca_deepmut/CNN_training/omicronBA1_adj_data.csv",header=None)
var_name = 'omicronBA1'
five_fold_splits(full_set,var_name)

# Run when data avalable
# full_set = pd.read_csv("/mnt/ncshare/ozkilim/charge_pca_deepmut/CNN_training/OmicronBA2_adj_data.csv",header=None)
# var_name = 'OmicronBA2'
# five_fold_splits(full_set,var_name)
```
